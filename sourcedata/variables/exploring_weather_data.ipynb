```python
# Calculate value counts
value_counts = wetter['Wettercode'].value_counts().sort_index()
print(value_counts)

# Set a frequency threshold (e.g., 10% of the total occurrences)
total_counts = value_counts.sum()
frequency_threshold = 0.10 * total_counts

# Select important weather codes
important_weather_codes = value_counts[value_counts >= frequency_threshold].index.tolist()
print(f"Important weather codes (threshold {frequency_threshold} occurrences): {important_weather_codes}")
```

```python
total_counts = value_counts.sum()

# Calculate cumulative frequency
cumulative_frequency = value_counts.cumsum() / total_counts

# Set a cumulative threshold (e.g., 90%)
cumulative_threshold = 0.90

# Select important weather codes
important_weather_codes = cumulative_frequency[cumulative_frequency <= cumulative_threshold].index.tolist()
print(f"Important weather codes (cumulative threshold {cumulative_threshold * 100}%): {important_weather_codes}")
```
